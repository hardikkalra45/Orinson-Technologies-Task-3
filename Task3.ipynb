```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
file_path = '/content/Customer Purchasing Behaviors.csv'
data = pd.read_csv(file_path)

# Selecting the feature and target variable
X = data[['purchase_frequency']]  # Feature
y = data['loyalty_score']  # Target

# Train a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Model performance metrics
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Plotting the regression line and data points
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred, color="red", label="Regression line")
plt.title("Linear Regression: Purchase Frequency vs Loyalty Score")
plt.xlabel("Purchase Frequency")
plt.ylabel("Loyalty Score")
plt.legend()
plt.show()

print(f"MSE: {mse}, R² Score: {r2}")
```
